```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# получение текущей директории
current_directory = os.getcwd()

# Путь к дир
image_directory = "/Users/mac/Documents/R_D/COM_VIS/Project/track/basetrack/blAd"

# Получ списка с дир
image_files = os.listdir(image_directory)

# Случайн изображения
np.random.shuffle(image_files)

# Отображаем кол-во 
num_images_to_display = 15

# Отображ случ изображений
plt.figure(figsize=(12, 8))
for i in range(num_images_to_display):
    # Загрузка 
    img_path = os.path.join(image_directory, image_files[i])
    img = cv2.imread(img_path)
    # Конвертация BGR в RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Отображение
    plt.subplot(3, 5, i+1)
    plt.imshow(img)
    plt.title(image_files[i])
    plt.axis('off')
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import os

# Шлях до кореневої директорії з папками класів
root_directory = "/Users/mac/Documents/R_D/COM_VIS/Project/track/basetrack/Ad"

# Отримання списку папок (класів)
class_folders = os.listdir(root_directory)

# Створення порожнього словника для зберігання кількості зображень для кожного класу
class_counts = {}

# Перегляд кожної папки
for class_folder in class_folders:
    # Шлях до папки класу
    class_path = os.path.join(root_directory, class_folder)
    # Перелік файлів у папці (зображення)
    images = os.listdir(class_path)
    # Підрахунок кількості зображень у кожній папці (класі)
    class_counts[class_folder] = len(images)

# Створення гістограми
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel('Класс')
plt.ylabel('Кол-во изобр')
plt.title('Гистограма распред классов')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# Путь с папкамим
root_directory = "/Users/mac/Documents/R_D/COM_VIS/Project/track/basetrack/Ad"

# Созд пустых списков
widths = []
heights = []

# Просмотр классов
for class_folder in os.listdir(root_directory):
    class_path = os.path.join(root_directory, class_folder)
    # Просмотр изображений
    for image_file in os.listdir(class_path):
        image_path = os.path.join(class_path, image_file)
        # Получение размера
        width, height = Image.open(image_path).size
        widths.append(width)
        heights.append(height)

# Постр гистограммы
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(widths, bins=20, color='blue', alpha=0.7)
plt.title('Ширина изображений')
plt.xlabel('Ширина')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(heights, bins=20, color='green', alpha=0.7)
plt.title('Высота изображений')
plt.xlabel('Высота')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()

```
