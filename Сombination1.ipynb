```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def draw_sphere():
    phi = np.linspace(0, np.pi, 100)
    theta = np.linspace(0, 2*np.pi, 100)
    phi_grid, theta_grid = np.meshgrid(phi, theta)
    x_grid = np.sin(phi_grid) * np.cos(theta_grid)
    y_grid = np.sin(phi_grid) * np.sin(theta_grid)
    z_grid = np.cos(phi_grid)
    ax.plot_surface(x_grid, y_grid, z_grid, alpha=0.5)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
draw_sphere()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_cylinder(ax, center, radius):
    z = np.linspace(-1, 1, 100)
    theta = np.linspace(0, 2*np.pi, 100)
    Z, Theta = np.meshgrid(z, theta)
    X = center[0] + radius * np.cos(Theta)
    Y = center[1] + radius * np.sin(Theta)
    ax.plot_surface(X, Y, Z, alpha=0.5)

def plot_sphere(ax, center, radius):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
    y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
    z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color='b', alpha=0.5)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Radius of the cylinder and sphere
cylinder_radius = 1.0

# Coordinates for cylinder center
cylinder_center = (0, 0, 0)

# Plot the cylinder
plot_cylinder(ax, cylinder_center, cylinder_radius)

# Plot the sphere inside the cylinder
plot_sphere(ax, cylinder_center, cylinder_radius)

# Show the plot
plt.show()

```
