```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread('001.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Create a blurred (unsharp) version of the original image using Gaussian blurring
sigma = 1.5  # You can adjust the sigma value for the Gaussian blur
unsharp = cv2.GaussianBlur(img, (0, 0), sigma)

# Create the difference image (original − unsharp)
diff = cv2.subtract(img, unsharp)

# Apply USM to get the resulting image using `sharpened = original + (original − unsharp) × amount`
amount = 2.0  # You can adjust the amount of sharpening
sharpened = cv2.addWeighted(img, 1 + amount, diff, -amount, 0)

# Display the images
plt.subplot(131), plt.imshow(img), plt.title('Original Image')
plt.subplot(132), plt.imshow(unsharp), plt.title('Unsharp Image')
plt.subplot(133), plt.imshow(sharpened), plt.title('Sharpened Image')

plt.show()
```
