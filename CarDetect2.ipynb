```python
import cv2

# Open the video file
cap = cv2.VideoCapture('/Users/mac/Documents/R_D/COM_VIS/Diplome/set/data/scr04.mov')

# Create background subtractor with adjusted parameters
fgbg = cv2.createBackgroundSubtractorMOG2(history=500, detectShadows=False)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Apply background subtraction
    fgmask = fgbg.apply(frame)

    # Get contours from detected motion
    contours, _ = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Define size thresholds for different object classes
    track_area_threshold = 1000  # Adjust as needed
    car_area_threshold = 800    # Adjust as needed

    # Iterate through contours and classify objects
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > track_area_threshold:
            # Classify as "track" and draw bounding box
            x, y, w, h = cv2.boundingRect(contour)
            cv2.putText(frame, 'track', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
        elif area > car_area_threshold:
            # Classify as "car" and draw bounding box
            x, y, w, h = cv2.boundingRect(contour)
            cv2.putText(frame, 'car', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2, cv2.LINE_AA)

    # Display the resulting frame
    cv2.imshow('Object Classification', frame)

    # Exit on 'q' key press
    if cv2.waitKey(15) & 0xFF == ord('q'):
        break

# Release video file and close image window
cap.release()
cv2.destroyAllWindows()
```
