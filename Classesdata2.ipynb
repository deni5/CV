```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

current_directory = os.getcwd()

root = os.path.join(current_directory, 'Train.csv')

data = pd.read_csv(root)

# Number of training samples (amount of samples in data)
num_samples = len(data)

# Show random data samples
for ii in range(15):
    # Get random index
    idx = np.random.randint(0, num_samples)
    # Load image
    img = cv2.imread(os.path.join(current_directory, data.iloc[idx]['Path'])) 
    if img is not None:
        # Convert image to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Show image
        plt.subplot(3,5,ii+1), plt.imshow(img), plt.title(data.iloc[idx]['Path'])
    else:
        print(f"no file {idx} ")

plt.show()
```

```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Train.csv')

# Extract class identifiers from the 'ClassId' column
ids = data['ClassId']

# Compute the histogram using Counter
hist = Counter(ids)

# Sort the classes by their counts in descending order
sorted_classes = sorted(hist.items(), key=lambda x: x[1], reverse=True)

# Display the most represented classes (top 5)
print("Most represented classes:")
for class_id, count in sorted_classes[:5]:
    print(f"Class ID: {class_id}, Count: {count}")

# Display the least represented classes (bottom 5)
print("\nLeast represented classes:")
for class_id, count in sorted_classes[-5:]:
    print(f"Class ID: {class_id}, Count: {count}")

# Plot the histogram
plt.bar(hist.keys(), hist.values())
plt.xlabel('Traffic Sign ID')
plt.ylabel('Counts')
plt.title('Per-Class Histogram')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract width and height data
widths = data['Width']
heights = data['Height']

# Plot histograms for width and height
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(widths, bins=20, color='blue', alpha=0.7)
plt.title('Width Distribution')
plt.xlabel('Width')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(heights, bins=20, color='green', alpha=0.7)
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract data for Roi.X1, Roi.Y1, Roi.X2, Roi.Y2
roi_x1 = data['Roi.X1']
roi_y1 = data['Roi.Y1']
roi_x2 = data['Roi.X2']
roi_y2 = data['Roi.Y2']

# Plot histograms for each characteristic
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.hist(roi_x1, bins=20, color='red', alpha=0.7)
plt.title('Roi.X1 Distribution')
plt.xlabel('Roi.X1')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(roi_y1, bins=20, color='orange', alpha=0.7)
plt.title('Roi.Y1 Distribution')
plt.xlabel('Roi.Y1')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(roi_x2, bins=20, color='purple', alpha=0.7)
plt.title('Roi.X2 Distribution')
plt.xlabel('Roi.X2')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(roi_y2, bins=20, color='pink', alpha=0.7)
plt.title('Roi.Y2 Distribution')
plt.xlabel('Roi.Y2')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
